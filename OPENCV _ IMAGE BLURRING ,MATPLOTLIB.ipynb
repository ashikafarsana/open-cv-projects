```python

import cv2
import numpy as np
import matplotlib.pyplot as plt
image=cv2.imread("C:/Users/PARCHU/Desktop/opencv/2022.jpg")
cv2.imshow("ORIGINAL IMAGE",image)
cv2.waitKey() 
cv2.destroyAllWindows()
plt.imshow(image)
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img2=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.xticks([]), plt.yticks([])
plt.imshow(img2)
plt.title("original image")
```

```python
Gaussian = cv2.GaussianBlur(image, (7, 7), 0) 
cv2.imshow("GAUSSIAN BLURR",Gaussian)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(Gaussian)
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img3=cv2.cvtColor(Gaussian,cv2.COLOR_BGR2RGB)
plt.xticks([]), plt.yticks([])
plt.imshow(img3)
plt.title("GAUSSIAN BLURR")
```

```python
median = cv2.medianBlur(image, 7) 
cv2.imshow('MEDIAN BLURRING', median) 
cv2.waitKey() 
cv2.destroyAllWindows()  
plt.imshow(median)
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img4=cv2.cvtColor(median,cv2.COLOR_BGR2RGB)
plt.xticks([]), plt.yticks([])
plt.imshow(img4)
plt.title("MEDIAN BLURR")
```

```python
bilateral=cv2.bilateralFilter(image,9,75,75)
cv2.imshow("BILATERAL BLURRING",bilateral)
cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(bilateral)
plt.xticks([]), plt.yticks([])
plt.show()
```

```python
img5=cv2.cvtColor(bilateral,cv2.COLOR_BGR2RGB)
plt.xticks([]), plt.yticks([])
plt.imshow(img5)
plt.title("BILATERAL BLURR")
```
